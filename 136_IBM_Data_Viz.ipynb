```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create list of years as strings
years = list(map(str, range(1980, 2014)))

# Create df_total with year and total immigrant count
df_total = pd.DataFrame({
    'year': years,
    'Total': df_canada[years].sum(axis=0).values
})

# Convert 'year' to numeric
df_total['year'] = df_total['year'].astype(int)

# Fit a linear regression model using numpy
x = df_total['year']
y = df_total['Total']
fit = np.polyfit(x, y, deg=1)  # degree=1 for linear

# Create polynomial function from coefficients
regression_line = np.poly1d(fit)

# Print the best-fit formula
print(f"Best fit line: y = {fit[0]:.2f} * x + {fit[1]:.2f}")

# Plot scatter
plt.figure(figsize=(10, 6))
plt.scatter(x, y, label='Data')

# Plot regression line
plt.plot(x, regression_line(x), color='black', linewidth=2, label='Best Fit Line')

# Labels and title
plt.title('Total Immigrant Population to Canada from 19802013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.legend()
plt.grid(True)
plt.show()

# Show the data frame
df_total
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define years
years = list(map(str, range(1980, 2014)))
year_axis = list(map(int, years))

# Extract data for Turkey and Germany
turkey = df_canada.loc['Turkey', years].astype(int)
germany = df_canada.loc['Germany', years].astype(int)

# Normalize bubble sizes using min-max scaling
turkey_normalized = (turkey - turkey.min()) / (turkey.max() - turkey.min())
germany_normalized = (germany - germany.min()) / (germany.max() - germany.min())

# Scale for bubble sizes (e.g., range from 50 to 300)
turkey_sizes = 50 + turkey_normalized * 250
germany_sizes = 50 + germany_normalized * 250

# Create the plot
plt.figure(figsize=(12, 6))

# Plot Turkey
plt.scatter(year_axis, turkey, 
            s=turkey_sizes,
            alpha=0.6,
            label='Turkey',
            color='orange',
            edgecolors='black')

# Plot Germany
plt.scatter(year_axis, germany, 
            s=germany_sizes,
            alpha=0.6,
            label='Germany',
            color='blue',
            edgecolors='black')

# Add labels and title
plt.title('Immigration to Canada: Turkey vs. Germany (19802013)')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Synthetic Data
years = np.arange(1980, 2014)
immigrants = np.random.randint(2000, 10000, size=(34,))

# Plot the scatter
ax.scatter(years, immigrants, label="Immigrants (scatter)", color='blue')

# Plot the line with square markers and dotted line
ax.plot(years, immigrants,
        marker='s',
        markersize=5,
        color='green',
        linestyle='dotted',
        label="Immigrants (line)")

# Legend at upper center of the figure
ax.legend(["Immigrants"], loc = 'upper center')

# Add title and labels
ax.set_title('Immigrants between 1980 to 2013')
ax.set_xlabel('Years')
ax.set_ylabel('Total Immigrants')

# Set x-axis limits
ax.set_xlim(1975, 2015)

# Enable grid
ax.grid(True)

# Add legend
ax.legend()

# Display the plot
plt.show()
```

```python
# create a figure with two axes in a row
fig, axs = plt.subplots(1,2,sharey=True)

# first plot (left)
axs[0].plot(years, immigrants)
axs[0].set_title("Line plot on immigrants")

# second plot (right)
axs[1].scatter(years, immigrants)
axs[1].set_title("Scatter plot on immigrants")
```
